```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.ensemble import IsolationForest, RandomForestRegressor, VotingRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest
from prophet import Prophet
from xgboost import XGBRegressor
import warnings

warnings.filterwarnings('ignore')

# ======================
# 1. Генерация датасета
# ======================
np.random.seed(42)
data_size = 2000  # Увеличили размер датасета

products = ["Эспрессо", "Американо", "Капучино", "Латте", "Раф",
           "Чай зеленый", "Чай черный", "Круассан классический",
           "Круассан миндальный", "Тарталетка лимонная",
           "Тарталетка с соленой карамелью", "Ролл с ветчиной и сыром",
           "Ролл с лососем и творожным сыром"]
prices = [3, 5, 8, 8, 10, 5, 5, 9, 11, 13, 13, 16, 19]

dates = pd.date_range('2024-01-01', '2025-05-06', periods=data_size)
times = [f"{np.random.randint(8, 22)}:{np.random.randint(0, 60):02d}" for _ in range(data_size)]

data = {
    'transaction_id': np.arange(1000, 1000 + data_size),
    'datetime': [d + pd.Timedelta(minutes=int(t.split(':')[1]), hours=int(t.split(':')[0]))
                for d, t in zip(dates, times)],
    'client_id': np.random.randint(1, 301, size=data_size),  # 300 уникальных клиентов
    'age': np.random.normal(loc=35, scale=10, size=data_size).astype(int).clip(16, 70),
    'gender': np.random.choice(['М', 'Ж'], size=data_size, p=[0.4, 0.6]),
    'product': np.random.choice(products, size=data_size, p=[0.15, 0.1, 0.2, 0.15, 0.05, 0.05, 0.05, 0.05, 0.03, 0.04, 0.03, 0.05, 0.05]),
    'quantity': np.random.choice([1, 2, 3, 4], size=data_size, p=[0.6, 0.25, 0.1, 0.05]),
    'price': [prices[products.index(p)] for p in np.random.choice(products, size=data_size, p=[0.15, 0.1, 0.2, 0.15, 0.05, 0.05, 0.05, 0.05, 0.03, 0.04, 0.03, 0.05, 0.05])],
    'payment_method': np.random.choice(['Карта', 'Наличные', 'Мобильный'], size=data_size, p=[0.6, 0.3, 0.1]),
    'loyalty_card': np.random.choice([True, False], size=data_size, p=[0.4, 0.6])
}
#расчет показателей
df = pd.DataFrame(data)
df['total_price'] = df['price'] * df['quantity']
df['day_of_week'] = df['datetime'].dt.day_name()
df['hour'] = df['datetime'].dt.hour
df['month'] = df['datetime'].dt.month_name()
df['is_weekend'] = df['day_of_week'].isin(['Saturday', 'Sunday'])
df['time_of_day'] = pd.cut(df['hour'], bins=[0, 11, 16, 24], labels=['Утро', 'День', 'Вечер'], include_lowest=True)

# ======================
# 2. Описание данных
# ======================
print("\n=== Описание колонок ===")
column_description = {
    'transaction_id': 'Уникальный идентификатор транзакции',
    'datetime': 'Дата и время покупки',
    'client_id': 'Уникальный идентификатор клиента',
    'age': 'Возраст клиента (16-70 лет)',
    'gender': 'Пол клиента (М/Ж)',
    'product': 'Название товара',
    'quantity': 'Количество товара в покупке (1-4)',
    'price': 'Цена за единицу товара',
    'payment_method': 'Способ оплаты',
    'loyalty_card': 'Наличие карты лояльности',
    'total_price': 'Общая сумма покупки',
    'day_of_week': 'День недели',
    'hour': 'Час покупки (8-22)',
    'month': 'Месяц покупки',
    'is_weekend': 'Покупка в выходной день',
    'time_of_day': 'Время дня (Утро/День/Вечер)'
}

for col, desc in column_description.items():
    print(f"{col}: {desc}")

# ======================
# 3. EDA Анализ
# ======================
print("\n=== EDA Анализ ===")

# 3.1 Общая информация
print("\nПервые 5 строк:")
display(df.head())

print("\nИнформация о датасете:")
print(df.info())

print("\nСтатистика по числовым колонкам:")
display(df.describe())

print("\nСтатистика по категориальным колонкам:")
display(df.describe(include=['object', 'bool']))

# 3.2 Проверка на пропуски
print("\nПропущенные значения:")
print(df.isnull().sum())

# 3.3 Визуализация распределений
plt.figure(figsize=(18, 12))

# Распределение продаж по товарам
plt.subplot(2, 2, 1)
df['product'].value_counts().plot(kind='bar')
plt.title('Распределение продаж по товарам', fontsize=12)
plt.xlabel('Товар', fontsize=10)
plt.ylabel('Количество продаж', fontsize=10)
plt.xticks(rotation=45, fontsize=8)

# Распределение по времени дня
plt.subplot(2, 2, 2)
df['time_of_day'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Распределение по времени дня', fontsize=12)
plt.ylabel('')  # Убираем label по оси Y для pie chart

# Распределение суммы покупок
plt.subplot(2, 2, 3)
sns.histplot(df['total_price'], bins=30, kde=True)
plt.title('Распределение суммы покупок', fontsize=12)
plt.xlabel('Сумма покупки (руб)', fontsize=10)
plt.ylabel('Количество покупок', fontsize=10)

# Распределение по способам оплаты
plt.subplot(2, 2, 4)
df['payment_method'].value_counts().plot(kind='bar')
plt.title('Способы оплаты', fontsize=12)
plt.xlabel('Способ оплаты', fontsize=10)
plt.ylabel('Количество покупок', fontsize=10)

plt.tight_layout()
plt.show()

# Дополнительные графики с разбивкой по месяцам
months = df['month'].unique()
n_months = len(months)

# Графики продаж по месяцам
fig, axes = plt.subplots(n_months, 1, figsize=(12, 3*n_months))
fig.suptitle('Продажи по месяцам', y=1.02)

for i, month in enumerate(months):
    month_data = df[df['month'] == month]
    sns.lineplot(data=month_data.groupby('day_of_week')['total_price'].sum().reset_index(),
                 x='day_of_week', y='total_price', ax=axes[i], marker='o')
    axes[i].set_title(f'{month}', fontsize=10)
    axes[i].set_xlabel('День недели', fontsize=8)
    axes[i].set_ylabel('Сумма продаж (руб)', fontsize=8)
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Графики продаж по времени дня по месяцам
fig, axes = plt.subplots(n_months, 1, figsize=(12, 3*n_months))
fig.suptitle('Продажи по времени дня по месяцам', y=1.02)

for i, month in enumerate(months):
    month_data = df[df['month'] == month]
    sns.barplot(data=month_data, x='time_of_day', y='total_price', ax=axes[i], estimator=sum, ci=None)
    axes[i].set_title(f'{month}', fontsize=10)
    axes[i].set_xlabel('Время дня', fontsize=8)
    axes[i].set_ylabel('Сумма продаж (руб)', fontsize=8)

plt.tight_layout()
plt.show()

# ======================
# 4. Предобработка данных
# ======================
print("\n=== Предобработка данных ===")

# 4.1 Кодирование категориальных признаков
cat_cols = ['product', 'payment_method', 'day_of_week', 'gender', 'month', 'time_of_day']
df_encoded = pd.get_dummies(df, columns=cat_cols)

# 4.2 Масштабирование числовых признаков
num_cols = ['age', 'quantity', 'price', 'total_price', 'hour']
scaler = StandardScaler()
df_encoded[num_cols] = scaler.fit_transform(df_encoded[num_cols])

# ======================
# 5. Анализ временных рядов (5 подходов)
# ======================
print("\n=== Анализ временных рядов ===")

# Подготовка данных
ts_data = df.set_index('datetime')['total_price'].resample('D').sum()

# 5.1 Декомпозиция временного ряда с улучшенной визуализацией
print("\n=== Декомпозиция временного ряда ===")
decomposition = seasonal_decompose(ts_data, model='additive', period=30)

# Создаем фигуру с 4 рядами
plt.figure(figsize=(14, 10))

# 1. Исходный ряд
plt.subplot(4, 1, 1)
plt.plot(decomposition.observed, label='Исходный ряд')
plt.title('Исходный временной ряд', fontsize=12)
plt.ylabel('Сумма продаж (руб)', fontsize=10)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)

# 2. Тренд - зависимость продаж от трендов
plt.subplot(4, 1, 2)
plt.plot(decomposition.trend, label='Тренд', color='orange')
plt.title('Трендовая составляющая', fontsize=12)
plt.ylabel('Сумма продаж (руб)', fontsize=10)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)

# 3. Сезонность - влияние сезона
plt.subplot(4, 1, 3)
plt.plot(decomposition.seasonal, label='Сезонность', color='green')
plt.title('Сезонная составляющая', fontsize=12)
plt.ylabel('Отклонение (руб)', fontsize=10)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)

# 4. Остатки - отклонение (пример - рядом какое-то мероприятие)
plt.subplot(4, 1, 4)
plt.plot(decomposition.resid, label='Остатки', color='red')
plt.title('Случайная составляющая (остатки)', fontsize=12)
plt.xlabel('Дата', fontsize=10)
plt.ylabel('Отклонение (руб)', fontsize=10)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.suptitle('Декомпозиция временного ряда продаж', y=1.02, fontsize=14)
plt.show()

# 5.2 Prophet
print("\n=== Модель Prophet ===")
df_prophet = ts_data.reset_index()
df_prophet.columns = ['ds', 'y']
train_size = int(len(df_prophet) * 0.8)
train, test = df_prophet.iloc[:train_size], df_prophet.iloc[train_size:]

model_prophet = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False,
    changepoint_prior_scale=0.05
)
model_prophet.fit(train)
future = model_prophet.make_future_dataframe(periods=len(test))
forecast_prophet = model_prophet.predict(future)

# Визуализация прогноза Prophet с разбивкой по месяцам
fig = model_prophet.plot(forecast_prophet)
ax = fig.gca()
ax.set_title('Прогноз Prophet', fontsize=12)
ax.set_xlabel('Дата', fontsize=10)
ax.set_ylabel('Сумма продаж (руб)', fontsize=10)
plt.show()

# 5.3 Exponential Smoothing
print("\n=== Exponential Smoothing ===")
model_ets = ExponentialSmoothing(
    ts_data,
    seasonal='add',
    seasonal_periods=7
).fit()
forecast_ets = model_ets.forecast(len(test))

# 5.4 Random Forest для временных рядов
print("\n=== Random Forest для временных рядов ===")
def create_features(df, label=None):
    df = df.copy()
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['dayofyear'] = df.index.dayofyear
    if label:
        df[label] = df[label]
    return df

ts_features = create_features(ts_data.to_frame('total_price'))
X = ts_features.drop('total_price', axis=1)
y = ts_features['total_price']
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]

model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
model_rf.fit(X_train, y_train)
forecast_rf = model_rf.predict(X_test)

# 5.5 Gradient Boosting
print("\n=== Gradient Boosting ===")
model_gb = GradientBoostingRegressor(n_estimators=100, random_state=42)
model_gb.fit(X_train, y_train)
forecast_gb = model_gb.predict(X_test)

# 5.6 XGBoost
print("\n=== XGBoost ===")
model_xgb = XGBRegressor(n_estimators=100, random_state=42)
model_xgb.fit(X_train, y_train)
forecast_xgb = model_xgb.predict(X_test)

# Визуализация всех прогнозов с разбивкой по месяцам
plt.figure(figsize=(16, 8))
plt.plot(ts_data.index, ts_data.values, label='Фактические данные', color='blue')
plt.plot(test['ds'], test['y'], label='Тестовые данные', color='green')
plt.plot(forecast_prophet['ds'], forecast_prophet['yhat'], label='Prophet', linestyle='--')
plt.plot(test['ds'], forecast_ets.values, label='ETS', linestyle='--')
plt.plot(X_test.index, forecast_rf, label='Random Forest', linestyle='--')
plt.plot(X_test.index, forecast_gb, label='Gradient Boosting', linestyle='--')
plt.plot(X_test.index, forecast_xgb, label='XGBoost', linestyle='--')

# Добавляем вертикальные линии для разделения месяцев
month_starts = pd.date_range(start=ts_data.index.min(), end=ts_data.index.max(), freq='MS')
for month_start in month_starts:
    plt.axvline(x=month_start, color='gray', linestyle=':', alpha=0.5)

plt.title('Сравнение моделей временных рядов', fontsize=14)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Сумма продаж (руб)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# ======================
# 6. Ансамбль моделей
# ======================
print("\n=== Ансамбль моделей ===")
ensemble = VotingRegressor([
    ('rf', model_rf),
    ('gb', model_gb),
    ('xgb', model_xgb)
])
ensemble.fit(X_train, y_train)
forecast_ensemble = ensemble.predict(X_test)

# Оценка качества
models = {
    'Prophet': forecast_prophet.iloc[-len(test):]['yhat'],
    'ETS': forecast_ets,
    'Random Forest': forecast_rf,
    'Gradient Boosting': forecast_gb,
    'XGBoost': forecast_xgb,
    'Ансамбль': forecast_ensemble
}

results = []
for name, pred in models.items():
    mae = mean_absolute_error(test['y'], pred[-len(test):])
    mape = mean_absolute_percentage_error(test['y'], pred[-len(test):]) * 100
    results.append([name, mae, mape])

metrics_df = pd.DataFrame(results, columns=['Model', 'MAE', 'MAPE'])
print("\nСравнение моделей:")
display(metrics_df.sort_values('MAE'))

# ======================
# 7. Поиск аномалий
# Создаем копию DataFrame, чтобы не изменять исходные данные
df_anomaly = df.copy()

print("\n=== Выявление аномалий в продажах ===")

# Создаем и обучаем модель IsolationForest
iso = IsolationForest(contamination=0.05, random_state=42)
anomaly_scores = iso.fit_predict(df_anomaly[['total_price', 'quantity']])
df_anomaly['anomaly'] = np.where(anomaly_scores == -1, 1, 0)  # 1 - аномалия, 0 - норма

# Разделяем данные на нормальные и аномальные
normal = df_anomaly[df_anomaly['anomaly'] == 0]  # Исправлено: правильное условие фильтрации
anomalies = df_anomaly[df_anomaly['anomaly'] == 1]

# Создаем график
plt.figure(figsize=(14, 8))

# 1. График нормальных продаж
plt.scatter(normal['datetime'], normal['total_price'],
            color='blue', alpha=0.6, s=40, label='Нормальные продажи')

# 2. Выделяем аномалии
plt.scatter(anomalies['datetime'], anomalies['total_price'],
            color='red', alpha=1, s=60, label='Аномалии')

# 3. Добавляем разметку по месяцам
month_starts = pd.date_range(start=df_anomaly['datetime'].min(),
                           end=df_anomaly['datetime'].max(),
                           freq='MS')
for month_start in month_starts:
    plt.axvline(x=month_start, color='gray', linestyle=':', alpha=0.5)

# 4. Настраиваем отображение
plt.title('Аномальные продажи кофейни', fontsize=16, pad=20)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Сумма чека (руб)', fontsize=12)
plt.legend(fontsize=10)

# 5. Добавляем сетку
plt.grid(True, linestyle='--', alpha=0.3)

# 6. Примеры аномалий с пояснениями (3 случайные аномалии)
if not anomalies.empty:
    sample_anomalies = anomalies.sample(min(3, len(anomalies)), random_state=42)
    for idx, row in sample_anomalies.iterrows():
        plt.annotate(f"{row['product']}\n{row['quantity']} шт. × {row['price']} руб\nИтого: {row['total_price']} руб",
                    xy=(row['datetime'], row['total_price']),
                    xytext=(10, 20), textcoords='offset points',
                    arrowprops=dict(arrowstyle='->'),
                    bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.7))

plt.tight_layout()
plt.show()

# Анализ аномалий
print("\n=== Статистика аномалий ===")
print(f"Всего продаж: {len(df_anomaly)}")
print(f"Обнаружено аномалий: {len(anomalies)} ({len(anomalies)/len(df_anomaly)*100:.1f}%)")

if not anomalies.empty:
    print("\nТоп-5 самых крупных аномалий:")
    display(anomalies.nlargest(5, 'total_price')[['datetime', 'product', 'quantity', 'total_price']])

    print("\nРаспределение аномалий по товарам:")
    display(anomalies['product'].value_counts().head(5))

# ======================
# 8. Визуализация результатов
# ======================
print("\n=== Визуализация результатов ===")

# 8.1 График ошибок ансамблевой модели
errors = test['y'] - forecast_ensemble[-len(test):]
plt.figure(figsize=(10, 5))
sns.histplot(errors, bins=30, kde=True)
plt.title('Распределение ошибок ансамблевой модели', fontsize=12)
plt.xlabel('Ошибка прогноза (руб)', fontsize=10)
plt.ylabel('Количество случаев', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# 8.2 Важность признаков для Random Forest
feature_importances = pd.Series(model_rf.feature_importances_, index=X.columns)
plt.figure(figsize=(12, 6))
feature_importances.sort_values().plot(kind='barh')
plt.title('Важность признаков (Random Forest)', fontsize=12)
plt.xlabel('Важность', fontsize=10)
plt.ylabel('Признак', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# 8.3 Сравнение метрик качества
plt.figure(figsize=(12, 6))
sns.barplot(data=metrics_df, y='Model', x='MAE', palette='viridis')
plt.title('Сравнение MAE моделей', fontsize=12)
plt.xlabel('Средняя абсолютная ошибка (MAE)', fontsize=10)
plt.ylabel('Модель', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
